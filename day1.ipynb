```python
subset = df.iloc[:, 0::3] # what's going on here?
subset # basically we are starting from column 0 then moving onto to the next 3 columns repeatedly 
```

```python
subset = df.iloc[:, 0:6:] # what's going on here?
subset # start from  column 0 to 5 
```

```python
subset = df.iloc[:, :6:2] # what's going on here?
subset # this seems a bit awkward and confusing but the idea is to get column 0 and 2 and 4
```

```python
subset = df.iloc[:, ::2] # what's going on here?
subset # get column 0 and count by 2 to get the rest?
```

```python
subset = df.iloc[:, ::] # what's going on here?
subset # gets all columns
```

```python
# Now that we know how to slice indices in our DataFrame, we should look into aggregating and grouping.
df.head(n=15)
```

```python
# Recall that occurred_on column is object type. Let's converted into a datetime type.
df['occurred_on'] = pandas.to_datetime(df['occurred_on'])
```

```python
df['occurred_on'] 
# works like charm!
```

```python
# Now let's the average magnitude per year
subset = df.iloc[:, [1,5]]
subset
```

```python
subset.groupby('occurred_on')['magnitude'].mean()
# this doesn't work since we aren't targeting the year but the entire date of a row  
```

```python
subset.groupby(subset['occurred_on'].dt.year)['magnitude'].mean()
# we have to use the dt function to target a specific date or time 
```

```python
# now let's find the average magnitude and depth per year.
# we simply add depth to the list by adding double bracket to avoid future warnings.
mean_Magnitude_Depth_PerYear = df.groupby(df['occurred_on'].dt.year)[['magnitude','depth']].mean()
mean_Magnitude_Depth_PerYear
```

```python
pandas.set_option('display.max_rows', None) # Display all rows

total_Earthquakes_PerYear = df.\
        groupby(df['occurred_on'].dt.year)\
        ['cause'].value_counts()
total_Earthquake_PerYear
```

```python
# We can see that there were some earthquakes caused by a nuclear explosion or regular explosion.
# We should only focus on natural earthquakes.
total_Earthquakes_PerYear = df.\
        groupby(df['occurred_on'].dt.year)\
        ['cause'].value_counts()
total_Earthquakes_PerYear

# reviews[reviews['stars'] > 3].groupby('business_id')['stars'].count()

# total_Earthquakes_PerYear = df[df['cause' == 'earthquake']].groupby(df['occurred_on'].dt.year)['cause'].count()
```
